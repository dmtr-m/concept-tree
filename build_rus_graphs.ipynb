```python

import rus_anaphora
import graph.graph as graph
```

```python
text = '''
    Мама мыла оконную раму. Из неё выпало стекло. Оно разбилось о пол.
    На полу спал наш пёс шарик. Он услышал звук бьющегося стекла. Шарик залаял на маму.
    Мама включила телевизор. На Первом Канале выступал Григорий Лепс.
    Недавно Правительство РФ присвоило ему звание Народного Артиста РФ.
'''

doc = rus_anaphora.nlp(text)
pairs = rus_anaphora.get_syntactic_relations(doc)
```

```python
graph1 = graph.Graph()

for pair in pairs:
    try:
        graph1.add_vertex(pair[0])
    except:
        pass

    try:
        graph1.add_vertex(pair[2])
    except:
        pass

    graph1.add_edge(pair[0], pair[2], pair[1], 0, 0)
```

```python
print("Graph structure:")
print(graph1)
print("\nDetailed representation:")
print(repr(graph1))

print("\nEdges:")
for concept in graph1.vertices.keys():
    print(f"Concept: {concept}")
    print(graph1.get_vertex_edges(concept))

print("\nVertices:")
for concept in graph1.vertices.keys():
    print(graph1.vertices[concept])
```
